## 예제 6-2-1 그래프 꾸미기

```python
%matplotlib inline
import matplotlib.pyplot as plt
import sys                                               ## 파이썬 엔진에 대한 정보를 관리하는 모듈을 사용한다. 

from matplotlib import font_manager, rc                  ## 폰트를 관리하는 함수와 설정 함수를 사용한다. 

if sys.platform  == 'darwin':                             ## MAC OS의 이름을 확인한다.
    path = '/Library/Fonts/AppleGothic.ttf'  
elif sys.platform == 'win32':                             ## Windows 이름을 확인한다.
    path = "c:/Windows/Fonts/malgun.ttf"
else:
    print('Unknown system... sorry~~~~') 
    
font_name = font_manager.FontProperties(fname=path).get_name()        ##  폰트가 있는지를 확인한다. 
rc('font', family=font_name)                                          ## 한글 폰트를 시각화 환경에 세팅한다. 
plt.rcParams['axes.unicode_minus'] = False                            ## 한글 처리시 마이너스 값을 처리할 수 있도록 조정한다.
```

```python
fig = plt.figure(figsize=(8,4))                         ## 하나의 캔버스 객체를 만든다. 
                                                        ## 캔버스의 사이즈를 figsize에 할당한다. 
```

```python
fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(2, 1, 1)                          ## 캔버스 내에 실제 그래프가 들어갈 AXES 객체를 두 개 만든다.
ax2 = fig.add_subplot(2, 1, 2)                          ## 2행 1열을 지정하고 실제 그래프 위치를 마지막 인자로 지정한다.
plt.show()                                              ## 이 그래프를 보면 좌표에 대한 표시만 보여준다 
```

```python
fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(1, 2, 1)                ## 수평으로 두 개의 그래프를 그리려면 하나의 행에 열을 두 개 지정한다. 
ax2 = fig.add_subplot(1, 2, 2)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(8,4))      ## nrows, ncols 매개변수에 지정해서 표시할 수 있다. 
plt.show()
```

```python
f1 = plt.figure(figsize=(8,4))
plt.title(" Figure  ")                           ## 그래프의 제목을 부여한다. 
plt.plot([1, 2, 3, 4], 'ro:')                     ## 그래프의 선 색상을 적색(r)과 두 점이 만나는 marker인 원을  표시한다. 

f2 = plt.gcf()                                    ## 동일한 캔버스롤 다시 불러온다. 
plt.plot([4, 6, 8, 10], 'bo:')                    ## 그래프의 선 색상을 파란색(b)과 두 점이 만나는 marker인 원을 표시한다. 
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(8,4))      ## 캔버스와 axes 객체를 만든다.
plt.title(" Figure Axes  ")                  ## 제목을 추가한다.                             
axes.plot([1, 2, 3, 4], 'ro:')               ## axes 객체를 사용해서 그래프를 그린다.               

#f2 = plt.gcf()                               ## 다시 켄버스를 가져오지 않아도 axes 변수를 사용할 수 있다. 
axes.plot([4, 6, 8, 10], 'bo:')
plt.show()
```

```python
x = np.arange(1,10)                  ## 9 개의 원소를 가진 1차원 배열을 만든다.
y = x **2                            ##  이 배열을 제곱한  1차원 배열을 만든다.
                                     ##  하나의 함수를 만드는 것과 동일하다 
```

```python
fig = plt.figure(figsize=(8,4))
plt.title("제목을 한글로 표시")         ## 한글로 제목을 만든다. 
plt.plot(x,y)                           ## 그래프를 그래면 원래 원소의 제곱한 값이 만들어진다.                      
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))
plt.title("라인 스타일의 plot ")
plt.plot(x,y, 'rs--')                          ## 문자열로 색상, 마커, 그리고 선스타일을 지정해서 그래프를 변경한다.
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))
plt.title("라인 스타일의 plot ")
plt.plot(x,y, color='r', marker='s', linestyle='--')        ## 색상, 마커, 선스타일을 매개변수에 인자로 전달해서 그린다. 
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))
plt.plot(x,y, c="r", lw=3,ls="--", marker="s", ms=6, mec="b", mew=3, mfc="r")  ## 선의 굵기와 마커의 크기 등을 조정한다.
plt.title("스타일 적용 예제")
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))
plt.title("x축, y축의 범위 설정")
plt.plot(x,y, c="r", lw=3,ls="--", marker="s", ms=6, mec="b", mew=3, mfc="r")
plt.xlim(0, 10)                          ## 좌표의 범위를 조정한다. 
plt.ylim(0, 100)
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))
plt.plot(x,y, c="r", lw=3,ls="--", marker="s", ms=6, mec="b", mew=3, mfc="r")
plt.title("x축과 y축의 tick설정")
plt.xticks(x)                    ## 좌표 내의 숫자의 값을 지정한다. 
plt.yticks(y)
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))
plt.plot(x,y, c="r", lw=3,ls="--", marker="s", ms=6, mec="b", mew=3, mfc="r")
plt.title("x축과 y축의 tick설정")
plt.xticks(x)
plt.yticks(y)
plt.grid(True)               ## 좌표 내의 그리드를 표시한다 
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))
plt.plot(x,y, c="r", lw=3,ls="--", marker="s", ms=6, mec="b", mew=3, mfc="r")
plt.title("x축과 y축의 tick설정")
plt.xticks(x)
plt.yticks(y)
plt.grid(True)
plt.xlabel(" x 레이블 ")           ##   좌표 축의 제목을 부여한다. 
plt.ylabel(" y 레이블 ")
plt.show()
```

```python
x = np.linspace(0,10)
fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(x,np.sin(x))                     ## 사인 그래프를 그린다. 
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(x,np.cos(x))                     ## 코사인 그래프를 그린다
plt.show()
```
